```python
import pandas as pd

df = pd.read_csv("../data/raw/pl_2022_23.csv", sep=";", encoding="latin-1")
```

```python
df = df[df["Comp"] == "Premier League"]
```

```python
df = df[df["Pos"] != "GK"]
```

```python
df.shape
```

```python
df = df.groupby("Player", as_index=False).agg({
    "Squad": "first",
    "Pos": "first",
    "Min": "sum",
    "90s": "sum"
}).merge(
    df.groupby("Player").sum(numeric_only=True),
    on="Player"
)
```

```python
df.shape
```

```python
df["Min"] = df["Min_y"]
df["90s"] = df["90s_y"]
```

```python
df = df.drop(columns=["Min_x","90s_x","Min_y","90s_y"])
```

```python
df[["Min","90s"]].head()
```

```python
scs_columns = [
    "Player","Squad","Pos","Min","90s",

    # Defensive impact
    "Tkl","Int","Blocks","Clr",

    # Pressing
    "TklDef3rd","TklMid3rd","TklAtt3rd","Recov",

    # Progression
    "PasProg","CarProg","CarPrgDist","Pas3rd",

    # Chance creation
    "PasAss","SCA","PPA","TB",

    # Possession retention
    "PasTotCmp%","CarMis","CarDis",

    # Activity
    "Touches","Carries","Rec"
]

df = df[scs_columns]
```

```python
df.head()
```

```python
df.shape
```

```python
per90_columns = [
    "Tkl","Int","Blocks","Clr",
    "TklDef3rd","TklMid3rd","TklAtt3rd","Recov",
    "PasProg","CarProg","CarPrgDist","Pas3rd",
    "PasAss","SCA","PPA","TB",
    "Touches","Carries","Rec"
]

for col in per90_columns:
    df[col + "_p90"] = df[col] / df["90s"]
```
