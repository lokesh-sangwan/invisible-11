```python
import pandas as pd

df = pd.read_csv("../data/raw/epl_2022-23.csv", sep=";", encoding="latin-1")
```

```python
df = df[df["Comp"] == "Premier League"]
```

```python
df = df[df["Pos"] != "GK"]
```

```python
df.shape
```

```python
df = df.groupby("Player", as_index=False).agg({
    "Squad": "first",
    "Pos": "first",
    "Min": "sum",
    "90s": "sum"
}).merge(
    df.groupby("Player").sum(numeric_only=True),
    on="Player"
)
```

```python
df.shape
```

```python
df["Min"] = df["Min_y"]
df["90s"] = df["90s_y"]
```

```python
df = df.drop(columns=["Min_x","90s_x","Min_y","90s_y"])
```

```python
df[["Min","90s"]].head()
```

```python
scs_columns = [
    "Player","Squad","Pos","Min","90s",

    # Defensive impact
    "Tkl","Int","Blocks","Clr",

    # Pressing
    "TklDef3rd","TklMid3rd","TklAtt3rd","Recov",

    # Progression
    "PasProg","CarProg","CarPrgDist","Pas3rd",

    # Chance creation
    "PasAss","SCA","PPA","TB",

    # Possession retention
    "PasTotCmp%","CarMis","CarDis",

    # Activity
    "Touches","Carries","Rec"
]

df = df[scs_columns]
```

```python
df.head()
```

```python
df.shape
```

```python
per90_columns = [
    "Tkl","Int","Blocks","Clr",
    "TklDef3rd","TklMid3rd","TklAtt3rd","Recov",
    "PasProg","CarProg","CarPrgDist","Pas3rd",
    "PasAss","SCA","PPA","TB",
    "Touches","Carries","Rec"
]

for col in per90_columns:
    df[col + "_p90"] = df[col] / df["90s"]
```

```python
df.shape
```

## Phase 4 – SCS Dimension Construction

```python
scs_dimensions = {
    "defensive_impact": ["Tkl_p90", "Int_p90", "Blocks_p90", "Clr_p90"],
    
    "pressing_ball_winning": ["TklDef3rd_p90", "TklMid3rd_p90", "TklAtt3rd_p90", "Recov_p90"],
    
    "ball_progression": ["PasProg_p90", "CarProg_p90", "CarPrgDist_p90", "Pas3rd_p90"],
    
    "chance_creation": ["PasAss_p90", "SCA_p90", "PPA_p90", "TB_p90"],
    
    "possession_retention": ["PasTotCmp%", "CarMis", "CarDis"],
    
    "off_ball_activity": ["Touches_p90", "Carries_p90", "Rec_p90"]
}
```

### Applying Minimum Minutes Threshold

```python
# Filter players who played at least 900 minutes
df = df[df["Min"] >= 900]

# Check new shape
df.shape
```

### Standardizing Player Positions

```python
# Keep only the first listed position
df["Pos"] = df["Pos"].str[:2]

# Check the result
df["Pos"].value_counts()
```

### Position-Based Percentile Normalization

```python
scs_metrics = [
    "Tkl_p90","Int_p90","Blocks_p90","Clr_p90",
    "TklDef3rd_p90","TklMid3rd_p90","TklAtt3rd_p90","Recov_p90",
    "PasProg_p90","CarProg_p90","CarPrgDist_p90","Pas3rd_p90",
    "PasAss_p90","SCA_p90","PPA_p90","TB_p90",
    "PasTotCmp%","CarMis","CarDis",
    "Touches_p90","Carries_p90","Rec_p90"
]
```

```python
for stat in scs_metrics:
    
    df[stat + "_pct"] = (
        df.groupby("Pos")[stat]
        .rank(pct=True) * 100
    )

[c for c in df.columns if "_pct" in c]    
```

### Handling Negative Possession Metrics

```python
# Invert metrics where lower values are better
df["CarMis_pct"] = 100 - df["CarMis_pct"]
df["CarDis_pct"] = 100 - df["CarDis_pct"]
```

### Constructing SCS Performance Dimensions

```python
# Defensive Impact
df["defensive_impact"] = df[
    ["Tkl_p90_pct","Int_p90_pct","Blocks_p90_pct","Clr_p90_pct"]
].mean(axis=1)

# Pressing & Ball Winning
df["pressing_ball_winning"] = df[
    ["TklDef3rd_p90_pct","TklMid3rd_p90_pct","TklAtt3rd_p90_pct","Recov_p90_pct"]
].mean(axis=1)

# Ball Progression
df["ball_progression"] = df[
    ["PasProg_p90_pct","CarProg_p90_pct","CarPrgDist_p90_pct","Pas3rd_p90_pct"]
].mean(axis=1)

# Chance Creation
df["chance_creation"] = df[
    ["PasAss_p90_pct","SCA_p90_pct","PPA_p90_pct","TB_p90_pct"]
].mean(axis=1)

# Possession Retention
df["possession_retention"] = df[
    ["PasTotCmp%_pct","CarMis_pct","CarDis_pct"]
].mean(axis=1)

# Off Ball Activity
df["off_ball_activity"] = df[
    ["Touches_p90_pct","Carries_p90_pct","Rec_p90_pct"]
].mean(axis=1)
```

### Shadow Contribution Score (SCS)

```python
scs_weights = {
    
    "DF": {
        "defensive_impact": 0.30,
        "pressing_ball_winning": 0.20,
        "ball_progression": 0.15,
        "chance_creation": 0.05,
        "possession_retention": 0.15,
        "off_ball_activity": 0.15
    },
    
    "MF": {
        "defensive_impact": 0.15,
        "pressing_ball_winning": 0.20,
        "ball_progression": 0.20,
        "chance_creation": 0.15,
        "possession_retention": 0.15,
        "off_ball_activity": 0.15
    },
    
    "FW": {
        "defensive_impact": 0.05,
        "pressing_ball_winning": 0.10,
        "ball_progression": 0.20,
        "chance_creation": 0.30,
        "possession_retention": 0.15,
        "off_ball_activity": 0.20
    }
}
```

```python
def calculate_scs(row):
    
    weights = scs_weights[row["Pos"]]
    
    score = (
        row["defensive_impact"] * weights["defensive_impact"] +
        row["pressing_ball_winning"] * weights["pressing_ball_winning"] +
        row["ball_progression"] * weights["ball_progression"] +
        row["chance_creation"] * weights["chance_creation"] +
        row["possession_retention"] * weights["possession_retention"] +
        row["off_ball_activity"] * weights["off_ball_activity"]
    )
    
    return score

df["SCS"] = df.apply(calculate_scs, axis=1)
```

### Player Leaderboard

```python
leaderboard = (
    df.sort_values("SCS", ascending=False)
    [["Player","Squad","Pos","SCS"]]
    .reset_index(drop=True)
)


leaderboard.insert(0, "Rank", leaderboard.index + 1)

leaderboard["SCS"] = leaderboard["SCS"].round(2)
```
