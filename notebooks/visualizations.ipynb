```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("../data/processed/epl_2022-23_scs-ready.csv")
df["SCS"] = df["SCS"].round(2)
```

```python
top_players = (
    df.sort_values("SCS", ascending=False)
    .head(20)
)

top_players[["Player","Squad","Pos","SCS"]]
```

```python
fig = px.bar(
    top_players,
    x="SCS",
    y="Player",
    color="Pos",
    orientation="h",
    title="Top 20 Players by Shadow Contribution Score",
    hover_data=["Squad", "Goals"]
)
```

```python
fig.update_layout(
    title_x=0.5,
    title_font=dict(size=22),

    xaxis=dict(
        title="Shadow Contribution Score",
        title_font=dict(size=16)
    ),

    yaxis=dict(
        title="Player",
        title_font=dict(size=16),
        autorange="reversed"
    ),

    height=750,
    bargap=0.2,

    legend_title_text="Position",

    hoverlabel=dict(
        bgcolor="white",
        font_size=12
    )
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="Goals",
    y="SCS",
    color="Pos",
    hover_data=["Player", "Squad"],
    title="Player Impact Map: Revealing Contribution Beyond Goals",
    labels={
        "Goals": "Goals Scored",
        "SCS": "Shadow Contribution Score"
    },
)
```

```python
radar_features = [
    "defensive_impact",
    "pressing_ball_winning",
    "ball_progression",
    "chance_creation",
    "possession_retention",
    "off_ball_activity"
]
```

```python
def plot_player_radar(player_name, df):
    
    player_data = df[df["Player"] == player_name].iloc[0]
    
    values = [player_data[feature] for feature in radar_features]
    
    # close the radar loop
    values += values[:1]
    
    labels = [
        "Defensive Impact",
        "Pressing & Ball Winning",
        "Ball Progression",
        "Chance Creation",
        "Possession Retention",
        "Off-Ball Activity"
    ]
    
    labels += labels[:1]
    
    fig = go.Figure()
    
    fig.add_trace(go.Scatterpolar(
        r=values,
        theta=labels,
        fill='toself',
        line=dict(color="royalblue", width=3),
        fillcolor="rgba(65,105,225,0.25)",
        hovertemplate="<b>%{theta}</b><br>Percentile: %{r:.1f}<extra></extra>"
    ))
    
    fig.update_layout(
    title=dict(
        text=f"{player_name} — Performance Profile",
        x=0.5,
        font=dict(size=22)
    ),

    polar=dict(
        domain=dict(x=[0.05, 0.95], y=[0.05, 0.95]),   # ← key fix
        radialaxis=dict(
            visible=True,
            range=[0, 100],
            tickfont=dict(size=11),
            gridcolor="lightgray"
        ),
        angularaxis=dict(
            tickfont=dict(size=12)
        )
    ),

    height=600,
    showlegend=False,

    hoverlabel=dict(
        bgcolor="white",
        font_size=12
    )
)
    
    fig.show()
```

```python
plot_player_radar("Leandro Trossard", df)
```
