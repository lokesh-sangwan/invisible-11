```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("../data/processed/epl_2022-23_scs-ready.csv")
df["SCS"] = df["SCS"].round(2)
```

```python
top_players = (
    df.sort_values("SCS", ascending=False)
    .head(20)
)

top_players[["Player","Squad","Pos","SCS"]]
```

```python
fig = px.bar(
    top_players,
    x="SCS",
    y="Player",
    color="Pos",
    orientation="h",
    title="Top 20 Players by Shadow Contribution Score",
    hover_data=["Squad", "Goals"]
)

fig.update_layout(
    yaxis=dict(autorange="reversed"),
    xaxis_title="Shadow Contribution Score",
    yaxis_title="Player",
    height=700,
    bargap=0.10,
    title_x = 0.5,
    hoverlabel=dict(
        bgcolor="white",
        font_size=12,
        font_family="Arial"
    )
)

fig.show()
```
