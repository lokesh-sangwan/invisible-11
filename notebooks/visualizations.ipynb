```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("../data/processed/epl_2022-23_scs-ready.csv")
```

```python
top_players = (
    df.sort_values("SCS", ascending=False)
    .head(15)
)

top_players[["Player","Squad","Pos","SCS"]]
```

```python
fig = px.bar(
    top_players,
    x="SCS",
    y="Player",
    color="Pos",
    orientation="h",
    title="Top 15 Players by Shadow Contribution Score",
    hover_data=["Pos"]
)

fig.update_layout(
    yaxis=dict(autorange="reversed"),
    xaxis_title="Shadow Contribution Score",
    yaxis_title="Player"
)

fig.show()
```
